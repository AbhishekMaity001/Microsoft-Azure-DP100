# Searching in CDF

- Search the CDF resource types
- Filter the search results

**Code Pattern**

`client.<cdf_resource_type>.search()`

### Fuzzy Search on one field

````python
client.assets.search(name="Norway",limit=5)
````

````python
client.time_series.search(name="North",limit=5)
````

````python
client.files.search(name="sheet")
````

Similarly for sequences and events

`client.sequences.search(name="some name")`

`client.events.search(description="some description")`

### Multi-field fuzzy search

````python
client.assets.search(query="Nor",limit=5)
````

### Exact search on one field (e.g. name, description etc)

````python
client.assets.search(filter={"name": "Canada"})
````

### Retrieve data

#### Retrieve Raw Data

````python
# Don't put a long datetime range, as it'll fetch all the raw datapoints, which can be huge for Raw datapoints
client.time_series.data.retrieve(id=239507005016003, end='now')
````

#### Retrieve Aggregated data

````python
client.time_series.data.retrieve_dataframe(id=239507005016003,
                    end='1000w-ago',
                    aggregates=["average","sum"],
                    granularity="1h")
````

Retrieve latest datapoint before a particular time (last or any past time point)

````python
client.time_series.data.retrieve_latest(id=239507005016003, before="5w-ago")[0]
````

### Plot the data points

````python
client.time_series.data.retrieve(id=239507005016003, end='1000w-ago').to_pandas().plot()
````
